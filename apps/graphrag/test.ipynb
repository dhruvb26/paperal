```python
# Importing required libraries
from pypdf import PdfReader

# creating a pdf reader object
reader = PdfReader('gusnet.pdf')

# printing number of pages in pdf file
print(len(reader.pages))

# Initialize an empty list to store text from each page
pdf_text = []

# Iterate through each page in the PDF
for i in range(len(reader.pages)):
    page = reader.pages[i]
    text = page.extract_text()
    pdf_text.append(text)

# Write the extracted text to a TXT file
with open('pdf_text.txt', 'w', encoding='utf-8') as txtfile:
    for i, page_text in enumerate(pdf_text):
        txtfile.write(f'Page {i+1}\n{page_text}\n\n')
```
